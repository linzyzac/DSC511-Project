````python
# Calculate total reviews and recommended true counts
pivot_table = df.groupBy('received_for_free').agg(
    {'recommended': 'count'}
).withColumnRenamed('count(recommended)', 'Total Reviews')

recommended_true = df.filter(df['recommended'] == True).groupBy('received_for_free').count().withColumnRenamed('count', 'Recommended (True)')

# Join the two DataFrames
pivot_table = pivot_table.join(recommended_true, 'received_for_free', 'left_outer')

# Calculate proportion
pivot_table = pivot_table.withColumn('Proportion Recommended', pivot_table['Recommended (True)'] / pivot_table['Total Reviews'])

# Show results
pivot_table.show()
````

### Pre-processing

````python
from pyspark.ml.feature import StringIndexer
from pyspark.sql import Row
from pyspark.sql import types as T

# Improved Mapping Function to Handle All Cases
def map_recommendation(value):
    if isinstance(value, str):
        value = value.strip().lower()  # Handle spaces and case issues
        if value in ["true", "1", "yes"]:   # Flexible matching for true
            return 1
        elif value in ["false", "0", "no"]: # Flexible matching for false
            return 0
    elif isinstance(value, bool):  # If values are directly boolean
        return 1 if value else 0
    return 0  # Default value for unexpected data

# Register UDF (User Defined Function)
from pyspark.sql import functions as F
from pyspark.sql.functions import udf

map_recommendation_udf = udf(map_recommendation, T.IntegerType())

# Apply mapping logic
sample_df2_mappping = sample_df2.withColumn("recommended", map_recommendation_udf("recommended"))
````

````python
# Check distribution after mapping
sample_df2_mappping.groupBy("recommended").count().show()
````

````python
# separating the classes
majority_class = sample_df2_mappping.filter("recommended = 1")
minority_class = sample_df2_mappping.filter("recommended = 0")

# downsampling the majority class
ratio = minority_class.count() / majority_class.count()
downsampled_majority = majority_class.sample(False, ratio, seed=42)

# combining the downsampled majority and the original minority
balanced_data = downsampled_majority.union(minority_class)
````

````python
# verifying class balance
balanced_data.groupBy("recommended").count().show()
````

````python
# Index user and item IDs for collaborative filtering
user_indexer = StringIndexer(inputCol="review_id", outputCol="userId").fit(balanced_data)
item_indexer = StringIndexer(inputCol="app_id", outputCol="itemId").fit(balanced_data)

# Apply transformations
balanced_data = user_indexer.transform(balanced_data)
balanced_data = item_indexer.transform(balanced_data)

# Select relevant columns
data = balanced_data.select("userId", "itemId", "recommended")
````

````python
data.show(5)
````

````python
data.groupBy('recommended').count().show()
````

### Collaborative Filtering Using ALS (Alternating Least Squares)

````python
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Define ALS model
als = ALS(
    maxIter=15,
    regParam=0.05,
    rank=30,
    userCol="userId",
    itemCol="itemId",
    ratingCol="recommended",
    coldStartStrategy="drop" # Prevent issues with unseen items in test data
)

# Train-test split
train_data, test_data = data.randomSplit([0.7, 0.3], seed=42) #changed it to 70-30.

# Train the model
model = als.fit(train_data)
````

````python
# Predictions
predictions = model.transform(test_data)

# Cast 'prediction' to double to satisfy evaluator's requirement
predictions = predictions.withColumn("prediction", predictions["prediction"].cast("double"))

# Evaluate using BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator(
    labelCol="recommended",
    rawPredictionCol="prediction",
    metricName="areaUnderROC"
)

# Compute AUC
auc = evaluator.evaluate(predictions)
print(f"AUC (Area Under Curve): {auc:.4f}")
````

### Cosine Similarity

````python
import numpy as np

# Extract item factors from ALS model
item_factors = model.itemFactors.rdd.map(lambda row: (row.id, np.array(row.features)))

# Compute cosine similarity using RDD
def cosine_similarity(x, y):
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    if norm_x == 0 or norm_y == 0:   # Handle zero vectors
        return 0
    return float(np.dot(x, y) / (norm_x * norm_y))

# Generate all possible combinations for similarity comparison
pairs = item_factors.cartesian(item_factors).filter(lambda x: x[0][0] != x[1][0])

# Compute cosine similarity for each pair
similarity_rdd = pairs.map(lambda pair: (
    pair[0][0],   # Item 1 ID
    pair[1][0],   # Item 2 ID
    cosine_similarity(pair[0][1], pair[1][1])  # Cosine similarity score
))

# Convert back to DataFrame
similarity_df = spark.createDataFrame(similarity_rdd, ["item1", "item2", "similarity"])

# Filter most similar items (e.g., similarity > 0.5)
top_similar_items = similarity_df.filter("similarity > 0.5") # similarity>0.8 returns 0 rows
top_similar_items.show(10)
````

````python
# Extract user factors from ALS model
user_factors = model.userFactors.rdd.map(lambda row: (row.id, np.array(row.features)))
user_factors = user_factors.filter(lambda x: np.linalg.norm(x[1]) != 0)

# Generate all possible user pairs
user_pairs = user_factors.cartesian(user_factors).filter(lambda x: x[0][0] != x[1][0])

# Compute cosine similarity for each pair
user_similarity_rdd = user_pairs.map(lambda pair: (
    pair[0][0],   # User 1 ID
    pair[1][0],   # User 2 ID
    cosine_similarity(pair[0][1], pair[1][1])  # Cosine similarity score
))

# Convert back to DataFrame
user_similarity_df = spark.createDataFrame(user_similarity_rdd, ["user1", "user2", "similarity"])

# Filter for strong similarity (e.g., > 0.8)
top_similar_users = user_similarity_df.filter("similarity > 0.8")
````

Since ALS' .recommendForAllUsers() is not available with Logistic Regression, we can compute cosine similarity between users and items directly.

````python
# Compute cosine similarity using RDD
def cosine_similarity(x, y):
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    
    # Handle zero vectors safely
    if norm_x == 0 or norm_y == 0:
        return 0.0  # Assign similarity as zero if either vector is zero
    
    return float(np.dot(x, y) / (norm_x * norm_y))
````

````python
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, FloatType
````

````python
# User-based

# Extract user features from trained model
user_factors_lr = train_data_lr.rdd.map(lambda row: (row.features[0], np.array(row.features)))

# Generate all possible user pairs
user_pairs_lr = user_factors_lr.cartesian(user_factors_lr).filter(lambda x: x[0][0] != x[1][0])

# Compute Cosine Similarity for Users
user_similarity_rdd_lr = user_pairs_lr.map(lambda pair: (
    int(pair[0][0]),   # user1 ID
    int(pair[1][0]),   # user2 ID
    float(cosine_similarity(pair[0][1], pair[1][1]))  # similarity
))
````

````python
from pyspark.sql.types import StructType, StructField, IntegerType, FloatType

schema = StructType([
    StructField("user1", IntegerType(), True),
    StructField("user2", IntegerType(), True),
    StructField("similarity", FloatType(), True)
])

# Create DataFrame
user_similarity_df_lr = spark.createDataFrame(user_similarity_rdd_lr, schema)

# Filter most similar users
top_similar_users_lr = user_similarity_df_lr.filter("similarity > 0.5")
top_similar_users_lr.show(10)
````

````python
# Item-based

# Extract item factors
item_factors_lr = train_data_lr.rdd.map(lambda row: (row.features[1], np.array(row.features)))

# Generate all possible item pairs
item_pairs_lr = item_factors_lr.cartesian(item_factors_lr).filter(lambda x: x[0][0] != x[1][0])

item_similarity_rdd_lr = item_pairs_lr.map(lambda pair: (
    int(pair[0][0]),  # Item 1 ID as integer
    int(pair[1][0]),  # Item 2 ID as integer
    float(cosine_similarity(pair[0][1], pair[1][1]))  # Cosine similarity as float
))
````

````python
# Convert back to DataFrame
item_similarity_df_lr = spark.createDataFrame(item_similarity_rdd_lr, ["item1", "item2", "similarity"])

# Filter most similar items
top_similar_items_lr = item_similarity_df_lr.filter("similarity > 0.8")
top_similar_items_lr.show(10)
````

### Generating Recmmendations from cosine similarity from log reg

User-based recommendations

- Finding their most similar users using the cosine similarity table.

- Recommending items those similar users liked.

````python
train_data_lr.select("features").show()
````

````python
from pyspark.sql.functions import col
from pyspark.ml.functions import vector_to_array

# Convert 'features' vector column to array
train_data_lr = train_data_lr.withColumn("features_array", vector_to_array("features"))

# Access individual components (e.g., userId = 0, itemId = 1)
train_data_lr = train_data_lr \
    .withColumn("userId_lr", col("features_array")[0]) \
    .withColumn("itemId_lr", col("features_array")[1])
````

````python
# Select users with positive recommendations (recommended = 1)
positive_interactions = train_data_lr.filter("recommended == 1").select("userId_lr", "itemId_lr")

# Join positive interactions with similar users
user_recommendations = (
    top_similar_users_lr
    .join(positive_interactions, top_similar_users_lr.user2 == positive_interactions.userId_lr)
    .select("user1", "itemId_lr", "similarity")
)

# Rank recommended items by similarity score
from pyspark.sql import Window
from pyspark.sql.functions import row_number

windowSpec = Window.partitionBy("user1").orderBy(F.desc("similarity"))
recommended_items = user_recommendations.withColumn("rank", row_number().over(windowSpec))

# Filter top 10 recommendations per user
top_recommended_items = recommended_items.filter("rank <= 10").select("user1", "itemId_lr")
top_recommended_items.show(10)
````

Item-based recommendations

- Identify items a user has interacted with.

- Find similar items (from the item-item cosine similarity table).

- Recommend the top-scoring similar items.

````python
# Items that users interacted with
user_interactions = train_data_lr.select("userId_lr", "itemId_lr")

# Join with item similarity
item_recommendations = (
    user_interactions
    .join(top_similar_items_lr, user_interactions.itemId_lr == top_similar_items_lr.item1)
    .select("userId_lr", "item2", "similarity")
)

# Rank recommended items by similarity score
recommended_items_by_item = item_recommendations.withColumn("rank", row_number().over(windowSpec))

# Filter top 10 recommendations
top_recommended_items_by_item = recommended_items_by_item.filter("rank <= 10").select("userId_lr", "item2")
top_recommended_items_by_item.show(10)
````

Combine Recommendations (Hybrid Approach)

Combining both user-based and item-based recommendations can improve results.

````python
combined_recommendations = (
    top_recommended_items.union(top_recommended_items_by_item)
    .distinct()
    .orderBy("userId_lr")
)

combined_recommendations.show(10)
````
